#### 1.1 Euler solver

Build a function called $deriv(x,t)$ that returns f(x,t). k will be declared as a global variables, defined outside the function $deriv(x,t)$. You can use the following template. Print f for x0=1, t=10., k=1.

```python
def deriv(x, t):
    global k
    return -k * x

k = 1

print("f = ", deriv(1, 10))
```

Build a simple Euler solver: define dt, and write an iterative procedure that at each time step
* 1) compute f(x,t)
* 2) increment x according to the Euler scheme : x(t+dt)=x(t)+dt*f(x,t)
* 3) keep in memory the new value of x(t)
* 3) time=time+dt

Of course, before that you must initialize time=0 and x(t=0)=1


plot the result for dt=0.1 and for time going from 0 to t=5

```python
dt = 0.1
time = 0
x = 1

# Creating a list to store the values
time_values = [time]
x_values = [x]

# Euler's Method
while time <= 5:
    f = deriv(x, time) # Computing for f(x,t)
    x += dt * f #Euler's Method
    time += dt #Time increase

    # Storing the values
    time_values.append(time)
    x_values.append(x)

# Graph
plt.plot(time_values, x_values)
plt.xlabel("Time ------------->")
plt.ylabel("x(t) ------------->")
plt.title("Euler's Method for dx/dt = -kx with dt=0.1")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

#The Analytical Results
display(Math(r"\text{The Equation:}"))
display(Math(r"\frac{dx}{dt} = -kx"))

display(Math(r"\text{Separation of variables and integration:}"))
display(Math(r"\int \frac{1}{x} \, dx = -k \int dt"))

display(Math(r"\text{The Integral:}"))
display(Math(r"\ln |x| = -kt + C")) #where C is a constant

display(Math(r"\text{Solving for } x:"))
display(Math(r"x = e^{C} e^{-kt} = x_0 e^{-kt}"))

# at x(t=0) = 1
# 0 = e^(C) => C = 0

display(Math(r"\text{Solution:}"))
display(Math(r"x(t) = e^{-kt}"))

k = 1
dt = 0.1
time = 0
x = 1

time_values = [time]
x_numerical = [x]

def deriv(x, t):
    global k
    return -k * x

while time <= 5:
    f = deriv(x, time)
    x = x + dt * f
    time += dt
    time_values.append(time)
    x_numerical.append(x)

# Analytical soln.
x_analytical = np.exp(-k * np.array(time_values))

# Difference
difference = np.array(x_numerical) - x_analytical

# Plotting the analytical and numerical solutions
plt.figure(figsize=(12, 6))
plt.plot(time_values, x_analytical, label="Analytical Solution", color='blue')
plt.plot(time_values, x_numerical, 'o-', label="Numerical Solution (Euler)", color='orange')
plt.xlabel("Time ----------------------------------->")
plt.ylabel("x(t) ----------------------------------->")
plt.title("Analytical vs Numerical Solution for dx/dt = -kx")
plt.legend()
plt.grid()
plt.show()

# The difference
plt.figure(figsize=(12, 6))
plt.plot(time_values, difference, color='red')
plt.xlabel("Time ----------------------------------->")
plt.ylabel("Difference ----------------------------------->")
plt.title("Difference Between Numerical and Analytical Solutions")
plt.grid()
plt.show()

#The error grows with the increase in time for Euler's Method.
```

## 2 Energy conservative system: orbital motion in 2D (2nd order solver)

We now solve an energy conservative system, with no, or complex solution. We consider the gravity equation of a body with mass body with mass M. The equation of motion is :

$\frac{dX^2 }{dt^2}=-\frac{GM}{r^2}\times\frac{X}{r}$


$\frac{dY^2 }{dt^2}=-\frac{GM}{r^2}\times\frac{Y}{r}$




Where G is the universal gravity constant and $r=(X^2+Y^2)^{1/2}$. This is a coupled system of 2 second order equations. Rewrite it as a system of 4 first order equations, using $Vx$ and $Vy$ : the X and Y component of velocity in cartesian coordinates.

The system of 4 first order equations is :




$\frac{dV_x^2 }{dt^2}=-\frac{GM}{r^2}\times\frac{X}{r}$

$\frac{dx}{dt}=V_x$

$\frac{dV_y^2 }{dt^2}=-\frac{GM}{r^2}\times\frac{Y}{r}$

$\frac{dy}{dt}=V_y$


We will take the Earth-Moon system (check mass values on internet, work in METRIC SYSTEM UNITS !!) We will consider that the initial position of the moon at time=0 is X=384000km , Y=0, and that its velocity is Vx=0, $Vy=+a \omega $, where $a$ is the inital Earth-Moon distance (called semi-major axis), and $\omega=(GM/a^3)^{1/2}$ is the orbital angular velocity. The orbital period of the Moon, in, in theory should be $T=2\pi/\omega$

Write a function called $gravderiv(X,Y,Vx,Vy,t)$ that returns the derivative of vector (X,Y,Vx,Vy)

```python
def gravderiv(X,Y,Vx,Vy,t):
  r = (X**2+Y**2)**(1/2)
  return [Vx,Vy,-G*M*X/(r**3),-G*M*Y/(r**3)]
```

## 3 Energy conservative system 2 : going to 4th order

Re-use the function that computes the derivative for  planetary motion (exercice 2). Now change the solver and code a Runge Kutta 4 scheme, that is 4th order. Put your code below.

```python
# computing the derivative for the Earth-Moon system under gravity
def gravity_deriv(t, state, G, M):
    X, Y, Vx, Vy = state
    r = np.sqrt(X**2 + Y**2)
    dXdt = Vx
    dYdt = Vy
    dVxdt = -G * M * X / r**3
    dVydt = -G * M * Y / r**3
    return np.array([dXdt, dYdt, dVxdt, dVydt])

#Runge Kutta 4 (RK4) method
def rk4_step(f, y, t, dt, G, M):
    k1 = dt * f(t, y, G, M)
    k2 = dt * f(t + dt / 2, y + k1 / 2, G, M)
    k3 = dt * f(t + dt / 2, y + k2 / 2, G, M)
    k4 = dt * f(t + dt, y + k3, G, M)
    return y + (k1 + 2 * k2 + 2 * k3 + k4) / 6
```

## 4 : Interest of implicit solver !! Free-Fall in a viscous fluid

We consider the motion of a particule with mass m, radius r free-faling in a fluid. The particle is inserted at altitude Z with 0 velocity initially.  It suffers the weight force ($Fg=-mg$) plus a friction force, the "Stokes Force", so that :

$F_S =-6\pi \mu r v$

with $\mu$ is the fluid dynamical viscosity (in $Pa S$) , r is the particle radius, and v is its velocity (positive when directed upward, negative when directed downward, so that the Z axis is directed upward by convention). Note that we will neglect here the buyancy forces (sum of pressure forces onto the ball).

We solve the system of equation $\frac{dZ}{dt}=?$ and $\frac{d V_Z }{dt}=?$

Write the system of equation we want to  solve. We call U the vector U(t)=(Z(t),$V_z(t)$)

```python
# System of equations for Z and V_z
display(Math(r"\frac{dZ}{dt} = V_z"))
display(Math(r"\frac{dV_z}{dt} = -g - \frac{6\pi \mu r}{m} V_z"))

# Vector form of the system
display(Math(r"\text{Now, } U(t) = \begin{pmatrix} Z(t) \\ V_z(t) \end{pmatrix}"))
display(Math(r"\frac{dU}{dt} = \begin{pmatrix} \frac{dZ}{dt} \\ \frac{dV_z}{dt} \end{pmatrix} = \begin{pmatrix} V_z \\ -g - \frac{6\pi \mu r}{m} V_z \end{pmatrix}"))
```

Write a function call Deriv_free_fall$(U,t)$ that returns $\frac{dU}{dt}$

```python
def Deriv_free_fall(U, t):
    Z, V_z = U  # Unpack the vector U into Z and V_z
    dZ_dt = V_z
    dV_z_dt = g - (6 * np.pi * mu * r / m) * V_z
    return np.array([dZ_dt, dV_z_dt])
```

Solve the motion of the particle, using $\mu=1\times 10^{-3} kg/m/s$ (dynamic viscosity of water) and for radius=1cm, and for material density $\rho=3000 Kg/m^3$. Plot Z(t) and Vz(t) using explicit Euler.

```python
g = 9.81  # gravitational acceleration (m/s^2)
mu = 1e-3  # dynamic viscosity of water (kg/ms)
rho = 3000  # density of the particle material (kg/m^3)
r = 0.01  # radius of the particle (m)

# mass of the particle
m = rho * (4/3) * np.pi * r**3

# Initial conditions
Z0 = 10.0  # Initial altitude (m)
V_z0 = 0.0  # Initial velocity (m/s)
U0 = np.array([Z0, V_z0])  # Initial state vector [Z, V_z]

# Time
dt = 0.01  # time step
T_end = 200.0  # end time
N_steps = int(T_end / dt)  # number of time steps

# array to store the soln.
time = np.linspace(0, T_end, N_steps)
solution = np.zeros((N_steps, 2))  # Array to store [Z, V_z] at each time step
solution[0] = U0

# explicit Euler method
for i in range(1, N_steps):
    U = solution[i - 1]
    solution[i] = U + dt * Deriv_free_fall(U, time[i - 1])

# Z(t) and V_z(t) values
Z_t = solution[:, 0]
V_z_t = solution[:, 1]

# Plotting
plt.figure(figsize=(12, 5))

# Z(t)
plt.subplot(1, 2, 1)
plt.plot(time, -1 * Z_t)
plt.xlabel('Time (s) ------------------>')
plt.ylabel('Altitude Z (m) ------------------>')
plt.title('Altitude Z(t) vs Time')
plt.grid()

# V_z(t)
plt.subplot(1, 2, 2)
plt.plot(time, V_z_t, color='red')
plt.xlabel('Time (s) ------------------>')
plt.ylabel('Velocity V_z (m/s) ------------------>')
plt.title('Velocity V_z(t) vs Time')
plt.grid()

plt.tight_layout()
plt.show()

V_z_t_accurate = V_z_t
Z_t_accurate = Z_t
time_accurate = time
```

What is the terminal velocity $V_T$ ?

```python
# terminal velocity
V_T = (m * g) / (6 * np.pi * mu * r)

print("Terminal Velocity V_T = ",V_T,"m/s")
```

What is the coupling timescale $\tau$ (time necessary for the particle to reach 1/2 of $V_T$)

```python
# Calculate coupling timescale tau
tau = (m / (6 * np.pi * mu * r)) * np.log(2)

# Display the coupling timescale
print("Coupling Timescale  =", tau ,"seconds")
```

In order to cure that problem, we want to switch to an IMPLICIT EULER METHOD (1st order).
Write the Euler implicit numerical scheme  ($U_{n+1}=$ a function of $U_{n+1}$ and $U_n$)

```python
g = 9.81  # gravitational acceleration (m/s^2)
mu = 1e-3  # dynamic viscosity of water (kg/ms)
rho = 3000  # density of the particle material (kg/m^3)
r = 0.01  # radius of the particle (m)
m = rho * (4/3) * np.pi * r**3 # mass of the particle

# Initial conditions
Z0 = 10.0  # Initial altitude (m)
V_z0 = 0.0  # Initial velocity (m/s)
U0 = np.array([Z0, V_z0])  # Initial state vector [Z, V_z]

# Time
dt = 0.01  # time step
T_end = 200.0  # end time
N_steps = int(T_end / dt)  # number of time steps

# array to store the soln.
time = np.linspace(0, T_end, N_steps)
solution = np.zeros((N_steps, 2))  # Array to store [Z, V_z] at each time step
solution[0] = U0

# Implicit Euler method
for i in range(1, N_steps):
    U_old = solution[i - 1]
    U_new = U_old
    for _ in range(5):
        U_new = U_old + dt * Deriv_free_fall(U_new, time[i])

    solution[i] = U_new

# Z(t) and V_z(t) values
Z_t = solution[:, 0]
V_z_t = solution[:, 1]

# Plotting
plt.figure(figsize=(12, 5))

# Z(t)
plt.subplot(1, 2, 1)
plt.plot(time, -1 * Z_t)
plt.xlabel('Time (s) ------------------>')
plt.ylabel('Altitude Z (m) ------------------>')
plt.title('Altitude Z(t) vs Time')
plt.grid()

# V_z(t)
plt.subplot(1, 2, 2)
plt.plot(time, V_z_t, color='red')
plt.xlabel('Time (s) ------------------>')
plt.ylabel('Velocity V_z (m/s) ------------------>')
plt.title('Velocity V_z(t) vs Time')
plt.grid()

plt.tight_layout()
plt.show()

V_z_t_accurate = V_z_t
Z_t_accurate = Z_t
time_accurate = time
```
